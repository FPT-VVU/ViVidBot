```python
import numpy as np

np.random.seed(2024)

def generate_left_skewed_values(min_val, max_val, alpha, beta, num_samples):
    values = []
    scale = max_val - min_val
    while len(values) < num_samples:
        # Generate a value from a beta distribution
        sample = np.random.beta(alpha, beta)
        # Scale and shift the value to the desired range
        sample = int(min_val + sample * scale)
        # Append if within the desired range (additional check)
        if min_val <= sample <= max_val:
            values.append(sample)
    return values

# Parameters
min_val = 5
max_val = 300
alpha = 1.5  # Controls the shape of the beta distribution
beta = 10   # A larger beta compared to alpha increases left skewness
num_samples = 65000  # Number of values you want to generate

# Generate the values
random_values = generate_left_skewed_values(min_val, max_val, alpha, beta, num_samples)
```

```python
import matplotlib.pyplot as plt

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(random_values, bins=30, color='#38bdf8', edgecolor='black')
plt.title('Distribution of Generated Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np

def generate_one_left_skewed_value(min_val, max_val, alpha, beta):
    scale = max_val - min_val
    while True:
        # Generate a value from a beta distribution
        sample = np.random.beta(alpha, beta)
        # Scale and shift the value to the desired range
        sample = int(min_val + sample * scale)
        # Check if it's within the desired range and return if true
        if min_val <= sample <= max_val:
            return sample

# Parameters
min_val = 5
max_val = 300
alpha = 2  # Controls the shape of the beta distribution
beta = 5   # A larger beta compared to alpha increases left skewness

# Generate one value
value = generate_one_left_skewed_value(min_val, max_val, alpha, beta)
print("Generated Value:", value)
print(min_val + ":" + (min_val + value))
```
