```python
import json

data = json.load(open("data/generated_keywords.json"))
```

```python
keywords = set()

for d in data.values():
  keywords.update(d)
```

```python
len(keywords)
```

```python
from underthesea import chunk, word_tokenize
```

```python
word_tokenize("Tôi là sinh_viên trường đại_học Bách_Khoa Hà_Nội")
```

```python
word_dist = {}

for kw in keywords:
  tokens = chunk(kw)
  # print(tokens)
  # break
  for tk in tokens:
    t = tk[0]
    kind = tk[1]
    if kind not in ["Np", "N", "Ny"]:
      continue
    t = t.replace("_", " ")
    if t in word_dist:
      word_dist[t] += 1
    else:
      word_dist[t] = 1
```

```python
s = sorted(word_dist.items(), key=lambda x: x[1], reverse=True)
s
```

```python
# save to csv
# weight,word

with open("data/word_dist.csv", "w") as f:
  f.write("weight,word\n")
  for w, c in s:
    f.write(f"{c},{w}\n")
```
