```python
import json
from datasets import load_dataset

dataset = load_dataset("json", data_files="data/6.vivid_instruct_65k_result.jsonl")[
  "train"
]
```

```python
import os
import shutil


shard_index = 0
shard_limit = 500
current_shard = []

for i, example in enumerate(dataset):
  current_shard.append({**example, "shard_id": shard_index})
  if len(current_shard) == shard_limit or i == len(dataset) - 1:
    with open(f"shards/shard_{shard_index}.jsonl", "w") as f:
      for example in current_shard:
        f.write(json.dumps(example) + "\n")
    shard_index += 1
    current_shard = []
    print(f"Shard {shard_index} saved")

shutil.make_archive("shards", "zip", "shards")
```
